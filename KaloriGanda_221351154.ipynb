```python
import pandas as pd 
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_object as go
```

```python
df = pd.read_csv('kalori_ganda.csv')
```

```python
df.head()
```

```python
# Scatter plot 
plt.figure(figsize=(8, 6))
plt.scatter(x='Umur', y='Kalori', data=df, color='blue')
plt.title('Age vs Calori Intake')
plt.xlabel('Umur')
plt.ylabel('Kalori Intake (kcal)')
plt.grid(True)
plt.show()
```

```python
# Scatter plot 
plt.figure(figsize=(8, 6))
plt.scatter(x='BB', y='Kalori', data=df, color='red')
plt.title('Age vs Calori Intake')
plt.xlabel('BB (kg)')
plt.ylabel('Kalori Intake (kcal)')
plt.grid(True)
plt.show()
```

```python
# Histogram 
plt.figure(figsize=(8,6))
sns.histplot(x='BB', data=df)
plt.title('Distribusi BB')
plt.show()
```
