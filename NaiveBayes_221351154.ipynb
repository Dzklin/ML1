```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import accuracy_score, confusion_matrix

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Diabetes1.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Bar Chart
plt.figure(figsize=(8, 6))
sns.countplot(x='Diet', hue='Diabetes', data=df)
plt.title('Jumlah pasien diabetes')
plt.xlabel('Diet Type')
plt.ylabel('Count')
plt.legend(title='Diabetes')
plt.show()
```

```python
# pie chart
keluarga_history = df['Family_History'].value_counts()
plt.figure(figsize=(8, 6))
keluarga_history.plot.pie(autopct='%1.1f%%', startangle=140, colors=['blue','red'])
plt.title(' Persentase pasien dengan riwayat keluarga')
plt.ylabel(' ')
plt.show()
```

```python
# Line chart
df_susun = df.sort_values(by='Age')
plt.figure(figsize=(8, 6))
plt.plot(df_susun['Age'], df_susun['Blood_Sugar_Level'], marker='H')
plt.title("Tren kadar gula darah berdasarkan usia")
plt.xlabel('Age')
plt.ylabel('Tingkat Kadar gula darah (mf/Dl)')
plt.grid(True)
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='BMI', y='Blood_Sugar_Level', hue='Diabetes', data=df)
plt.title('Hubungan antara BMI dan Kadar Gula Darah')
plt.xlabel('BMI')
plt.ylabel('Blood Sugar Level')
plt.grid(True)
plt.show()
```
